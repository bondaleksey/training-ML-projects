```python
import pickle
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
filename = "../data/mathnet_bondarenko_page.pkl"
with open(filename,'rb') as inp:
    html = pickle.load(inp)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
print(soup.table)
```

```python
pubs= soup.table.find_all('td', attrs={'align':'left', 'valign':"top"})
print(pubs)
print(type(pubs))
```

```python
pubs= soup.find_all('td', attrs={'align':'left', 'valign':"top"})
print(pubs)
print(type(pubs))
```

```python
def get_paper_info(pub):
    result = {}
    name = re.sub('[\n\t\xa0]','',pub.text)
    name = re.sub(' +',' ',name)
    name = name.rstrip(' ')
    name = name.lstrip(' ')    
    result['name'] = name      
    for a in pub.find_all('a'):               
        link = a.get('href')
        if 'elibrary.ru' in link:
            # result['elib_link'] = link
            result['elib_id'] = link.replace("https://elibrary.ru/item.asp?id=","")
            # re.findall('\d+',link)[0]
        if 'mathnet.ru' in link:
            result['mn_link'] = link.replace("http://mi.mathnet.ru/","")
    return result
```

```python
papers={}
for indx, pub in enumerate(pubs[1:]):         
    papers[indx+1] = get_paper_info(pub)    
```

```python
print(papers)
```

```python
papers_dataframe = pd.DataFrame.from_dict(papers,orient='index')
```
