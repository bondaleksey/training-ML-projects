```python
import pickle
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
filename = "../data/mathnet_all_authors_page.pkl"
with open(filename,'rb') as inp:
    html = pickle.load(inp)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup)
```

```python
len(soup.prettify())
```

```python
fio=soup.find_all('a', attrs={'class':'SLink'})
# print(soup.find_all('a', attrs={'class':'SLink'}))
print((fio[31:40]))
print(len(fio[31:40]))
```

```python
# regex = r"personid=\d+\">\D+</"
personid = r"\d+"
# s = re.findall(regex,fio[31:40])
for line in fio[-10:-5]:
    # print(type(line['href']))
    print(line['href'])
    # print(type(line.text))
    print(line.text)
    print(re.findall(personid,line['href'])[0])
```

## Let's make authors dictionary 

```python
authors = {}
for ind, line in enumerate(fio[31:-5]):
    authors[ind+1]={
        'name':line.text,
        'mn_id':re.findall(personid,line['href'])[0]
    }
```

```python
# print(authors)
iam_authors_in_mathnet = pd.DataFrame.from_dict(authors,orient='index')
```

```python
iam_authors_in_mathnet.head(20)
```

```python
filename = "../data/mathnet_iam_authors_dict.pkl"
with open(filename,'wb') as outp:
    pickle.dump(authors, outp, pickle.HIGHEST_PROTOCOL)
```
