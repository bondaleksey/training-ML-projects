```python
import pickle
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
filename = "../data/one_article_example.pkl"
with open(filename,'rb') as inp:
    html = pickle.load(inp)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
s = soup.find_all('td',  attrs={'valign':"top"})
```

```python
print(len(s))
print(s)
```

```python
for line in s:    
    if "Аннотация:" in line.text:
        print(len(line.text))
        print(line)    
        mystr = line.text    
```

```python
print(type(mystr))
print(len(mystr))
```

```python
some = re.search(r'Ключевые(.*)\.', mystr)
print(some.group(1))
```

```python
# regex = r"<b></b>\.+<br/><br/>"
start_an = mystr.find("Аннотация") + len("Аннотация:")
start_kw = mystr.find("Ключевые слова:") + len("Ключевые слова:")
end_an = mystr.find("Ключевые слова:")
end_kw = mystr.index(".",start_kw )
print(mystr.find("Ключевые слова:"))
print(f"start_an = {start_an}")
print(f"end_an = {end_an}")
print(f"start_kw = {start_kw}")
print(f"end_kw = {end_kw}")
annotation = mystr[start_an:end_an]
key_words = mystr[start_kw:end_kw]
```
