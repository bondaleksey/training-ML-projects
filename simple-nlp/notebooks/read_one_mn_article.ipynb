```python
def read_article_file(au_mn_id, mn_paper_link):
    print("http://mi.mathnet.ru/"+mn_paper_link)
    filename = "../data/"+au_mn_id+"/"+mn_paper_link+".pkl"
    if exists(filename):
        print("we read from file")
        with open(filename,'rb') as inp:
            author_page = pickle.load(inp)
    else:
        page_link = "http://mi.mathnet.ru/"+mn_paper_link
        headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/28.0.1467.0 Safari/537.36'}
        
        # page_link ="http://mi.mathnet.ru/+ {}"
        # time.sleep(4*(1+rand))
        response = requests.get(page_link, headers=headers, timeout=None)
        if response.status_code == 200:
            author_page = response.content        
            save_html(author_page, au_mn_id, mn_paper_link,status='write')        
    return author_page
```
